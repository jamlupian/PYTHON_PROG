```python
"""
Task:
Leetspeak Translator (Parametric). Generalize our leetspeak idea: write translate(text:
str, mapping: dict[str,str]) -> str that applies a mapping (e.g., {’a’:’4’,’e’:’3’,...})
over lowercase letters only, leaving others unchanged. Explain why strings are immutable and show an efficient approach (e.g., list join vs. repeated concatenation).
Anchor concepts: replace, iteration strategies, efficiency. Reference: :contentReference[oaicite:7]index=7

Prompt:
Write a Python function translate(text, mapping) that replaces letters in the text based on the given dictionary (like leetspeak), applying the mapping only for lowercase letters and leaving all others unchanged. The function should return the new translated string.

Chatgpt Code:
"""
def translate(text, mapping):
    result = ""
    for ch in text:
        if ch in mapping:
            result += mapping[ch]
        else:
            result += ch
    return result

leet = {'a': '4', 'e': '3', 'i': '1', 'o': '0', 's': '5'}
print(translate("hello world", leet))
print(translate("LEETCODE", leet))
print("")
"""
Critique:

Correctness-
The code correctly goes through each character in the string and replaces it if it’s found in the given mapping.

Time & Space Complexity-
The code uses a loop to go through every character, so it runs in O(n) time where n is the length of the input string.

Robustness-
It works fine for normal strings, empty input, and even symbols or numbers. There’s no error handling needed because it doesn’t use risky operations.

Readability/Style-
The function is easy to read and understand. It follows clear naming and simple logic.

Faithfulness to Lecture-
This solution demonstrates the concept of string immutability well — the problem highlights why concatenating inside a loop is inefficient.

Improved Code:
"""
def translate(text: str, mapping: dict[str, str]) -> str:
    """
    Translate the input text using a character mapping (e.g., leetspeak).
    Only lowercase letters are replaced; others remain unchanged.
    """
    translated_chars = [
        mapping.get(ch, ch) if ch.islower() else ch
        for ch in text
    ]
    return ''.join(translated_chars)

if __name__ == "__main__":
    leet = {'a': '4', 'e': '3', 'i': '1', 'o': '0', 's': '5'}
    assert translate("hello world", leet) == "h3ll0 w0rld"
    assert translate("AEIOU", leet) == "AEIOU"
    assert translate("Pass123!", leet) == "P455123!"
    assert translate("", leet) == ""
    print(translate("hello world.", leet))
    print(translate("Leetspeak!", leet))
```
