```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
X, y_true = make_blobs(
    n_samples=500,
    centers=4,
    cluster_std=0.60,
    random_state=42
)

X[:5]  # preview
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)

y_kmeans = kmeans.predict(X)
centers = kmeans.cluster_centers_

y_kmeans[:10]  # preview cluster assignments
```

```python
plt.figure(figsize=(8, 6))

plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=30, cmap='viridis')
plt.scatter(
    centers[:, 0], centers[:, 1],
    c='red', s=200, marker='X', alpha=0.8
)

plt.title("K-Means Clustering on Synthetic make_blobs Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```
